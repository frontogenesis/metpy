```python
import xarray as xr

from metpy.cbook import get_test_data
from metpy.plots import ContourPlot, MapPanel, PanelContainer
from metpy.units import units
```

```python
narr = xr.open_dataset(get_test_data('narr_example.nc', as_file_obj=False))
```

```python
temperature_contours_850 = ContourPlot()
temperature_contours_850.data = narr
temperature_contours_850.field = 'Temperature'
temperature_contours_850.level = 850 * units.hPa
temperature_contours_850.linecolor = 'tab:red'
temperature_contours_850.contour = 15
```

```python
temperature_contours_500 = ContourPlot()
temperature_contours_500.data = narr
temperature_contours_500.field = 'Temperature'
temperature_contours_500.level = 500 * units.hPa
temperature_contours_500.linecolor = 'tab:red'
temperature_contours_500.contour = 15
```

```python
height_contours_850 = ContourPlot()
height_contours_850.data = narr
height_contours_850.field = 'Geopotential_height'
height_contours_850.level = 850 * units.hPa
height_contours_850.linecolor = 'black'
height_contours_850.contour = 15
```

```python
height_contours_500 = ContourPlot()
height_contours_500.data = narr
height_contours_500.field = 'Geopotential_height'
height_contours_500.level = 500 * units.hPa
height_contours_500.linecolor = 'black'
height_contours_500.contour = 15
```

```python
left_panel = MapPanel()
left_panel.area = 'us'
left_panel.layers = ['coastline', 'states']
left_panel.title = '850 hPa Map'
left_panel.plots = [temperature_contours_850, height_contours_850]
left_panel.layout = (1, 2, 1)
```

```python
right_panel = MapPanel()
right_panel.area = 'us'
right_panel.layers = ['coastline', 'states']
right_panel.title = '500 mb Map'
right_panel.plots = [temperature_contours_500, height_contours_500]
right_panel.layout = (1, 2, 2)
```

```python
pc = PanelContainer()
pc.size = (15, 5)
pc.panels = [left_panel, right_panel]
pc.show()
```
