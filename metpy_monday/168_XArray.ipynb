```python
import xarray
import numpy as np
import pandas as pd
```

```python
latitude = np.arange(20, 50.5, 0.5)
longitude = np.arange(-125, -65.5, 0.5)

print(np.shape(latitude), np.shape(longitude))
```

```python
levels = np.array([1000, 850, 500, 250])
print(np.shape(levels))
```

```python
times = pd.date_range(start='2/23/21 00:00:00', end='2/23/21 23:00:00', freq='H')
print(np.shape(times))
```

```python
temperature = np.random.random((24, 4, 61, 119))
print(np.shape(temperature))
```

```python
heights = np.array([0, 500, 1000, 1500, 2000])
print(np.shape(heights))
```

```python
wind_speed = np.random.random((24, 5, 61, 119))
print(np.shape(wind_speed))
```

```python
temperature_da = xarray.DataArray(temperature, coords=(times, levels, latitude, longitude),
                                  dims=['time', 'level', 'latitude', 'longitude'])

temperature_da.name = 'Air Temperature'
temperature_da.attrs['units'] = 'degC'
```

```python
wind_speed_da = xarray.DataArray(wind_speed, coords=(times, heights, latitude, longitude),
                                 dims=['time', 'level', 'latitude', 'longitude'])

wind_speed_da.name = 'Wind Speed'
wind_speed_da.attrs['units'] = 'knots'
```

```python
wind_speed_da
```

```python
ds = xarray.Dataset({'air_temperature': (['time', 'level', 'latitude', 'longitude'], temperature_da),
                     'wind_speed': (['time', 'height', 'latitude', 'longitude'], wind_speed_da)},
                    coords={'time': times, 'level': levels, 'height': heights,
                            'latitude': latitude, 'longitude': longitude})

ds['attrs'] = 'My Custom Model'
```

```python
ds
```

```python
single_time_temps = ds['air_temperature'].sel(time='2021-02-23 13:00:00', level=500)
```
