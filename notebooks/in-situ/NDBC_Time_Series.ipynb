```python
import matplotlib.pyplot as plt
from siphon.simplewebservice.ndbc import NDBC

plt.style.use('fivethirtyeight')
```

```python
data_aval = NDBC.buoy_data_types('41048')
```

```python
df = NDBC.realtime_observations('41048')
df = df.iloc[50:150]
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize=(12,10))

ax1.title.set_text('Buoy Time Series Plot - 41048')

# Pressure
ax1.plot(df['time'], df['pressure'], color='black')
ax1.set_ylabel('Pressure [hPa]')

# Wind speed, gust, direction
ax2.plot(df['time'], df['wind_speed'] * 2.23694, color='tab:orange')
ax2.plot(df['time'], df['wind_gust'] * 2.23694, color='tab:olive', linestyle='--')
ax2.set_ylabel('Wind Speed [mph]')

# Water Temperature
ax3.plot(df['time'], df['water_temperature'] * 9/5 + 32, color='tab:brown')
ax3.set_ylabel('Water Temp [degF]')

plt.savefig('41048.png')
```
